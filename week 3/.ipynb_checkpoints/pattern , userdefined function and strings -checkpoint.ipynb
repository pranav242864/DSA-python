```python
#using for loop :
n = 4

for i in range(1, n + 1):
    for j in range(1, n - i + 1):
        print(" ", end=" ")
    for k in range(i, 2 * i):
        print(k, end=" ")
    print()
```

```python
# Program to print Fibonacci series :
def fibonacci(n):
    a = 0
    b = 1
    for i in range(n):
        print(a, end=" ")
        c = a + b
        a = b
        b = c
n = int(input("Enter number of terms: "))
fibonacci(n)
```

```python
# Program to find factorial :
def factorial(n):
    fact = 1
    for i in range(1, n + 1):
        fact = fact * i
    print("Factorial:", fact)
n = int(input("Enter a number: "))
factorial(n)
```

```python
# Program to count vowels, consonants, digits, and special characters
def count_chars(s):
    vowels = 0
    consonants = 0
    digits = 0
    special = 0
    for ch in s:
        if ch.isalpha():
            if ch.lower() in "aeiou":
                vowels += 1
            else:
                consonants += 1
        elif ch.isdigit():
            digits += 1
        else:
            special += 1
    print("Vowels:", vowels)
    print("Consonants:", consonants)
    print("Digits:", digits)
    print("Special Characters:", special)
s = input("Enter a string: ")
count_chars(s)
```

```python
# Program to remove consecutive duplicate characters :
def remove_consecutive_duplicates(s):
    result = ""
    for ch in s:
        if result == "" or ch != result[-1]:
            result += ch
    return result
s = input("Enter a string: ")
print(remove_consecutive_duplicates(s))
```

```python
# Program to replace a character in a string :
def replace_char(s, old, new):
    result = ""
    for ch in s:
        if ch == old:
            result += new
        else:
            result += ch
    return result
s = input("Enter a string: ")
old = input("Enter character to replace: ")
new = input("Enter new character: ")
print(replace_char(s, old, new))
```
