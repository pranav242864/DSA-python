```python
s = input("Enter a string: ")
d = {}
for ch in s:
    if ch in d:
        d[ch] += 1
    else:
        d[ch]=1
print(d)
```

```python
# Store employee names and salaries in a dictionary
n = int(input("Enter number of employees: "))
employees = {}
for i in range(n):
    name = input("Enter employee name: ")
    salary = float(input("Enter salary: "))
    employees[name] = salary
print("Employee Details:")
print(employees)
```

```python
# Count occurrences of a character in a string
s = input("Enter a string: ")
ch = input("Enter a character to count: ")
count = 0
for c in s:
    if c == ch:
        count += 1
print("Character appears", count, "times")
```

```python
# Function to convert number into words (digit-wise)
def number_to_words(num):
    words = {
        '0': 'Zero',
        '1': 'One',
        '2': 'Two',
        '3': 'Three',
        '4': 'Four',
        '5': 'Five',
        '6': 'Six',
        '7': 'Seven',
        '8': 'Eight',
        '9': 'Nine'
    }
    for digit in num:
        print(words[digit], end=" ")
# User input
n = input("Enter a number: ")
number_to_words(n)
```
