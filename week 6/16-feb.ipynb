```python
class student :
    passingpercentage = 40
    def studentdetails(self):
        self.name = "pranav"
        print("name = ", self.name)
        self.percentage = 80
        print("percentage = ",self.percentage)

    def ispassed(self):
        if self.percentage > student.passingpercentage:
            print("passed")
        else :
            print("faild")
s1 = student()
s1.studentdetails()
s1.ispassed()
student.studentdetails(s1)
```
