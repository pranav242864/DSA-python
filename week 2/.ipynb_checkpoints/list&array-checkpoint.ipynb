```python
# Rotate an array by K positions (right rotation)
lst = [1, 2, 3, 4, 5]
k = int(input("Enter K: "))
k = k % len(lst)
rotated = lst[-k:] + lst[:-k]
print("Rotated array:", rotated)
```

```python
# Program to analyze body temperature and give advice (with out-of-range handling)
temp = float(input("Enter body temperature: "))
if temp >= 85.0 and temp <= 91.0:
    print("Serious Hypothermia")
elif temp > 91.0 and temp < 95.0:
    print("Mild Hypothermia")
elif temp >= 95.0 and temp <= 98.0:
    print("Normal Temperature")
elif temp > 98.0 and temp <= 100.0:
    print("Mild Fever")
elif temp > 100.0 and temp <= 105.0:
    print("High Fever")
else:
    print("Temperature out of human body range")
```

```python
# Program to check number type including zero (neutral)
n = int(input("Enter a number: "))
if n == 0:
    print("0 is a Neutral number")
elif n > 0 and n % 2 != 0:
    print("Number is Positive and Odd")
elif n > 0 and n % 2 == 0:
    print("Number is Positive and Even")
elif n < 0 and n % 2 != 0:
    print("Number is Negative and Odd")
elif n < 0 and n % 2 == 0:
    print("Number is Negative and Even")
```

```python
# to print number from 1 to n 
n = int(input("Enter N: "))
for i in range(1, n + 1):
    print(i, end=" ")
```

```python
# Take input as number, convert to string, then print digits right to left
n = int(input("Enter a number: "))
s = str(n)
for ch in reversed(s):
    print(ch)
```

```python
# Program to print all factors of a given positive number
n = int(input("Enter a positive number: "))
for i in range(1, n + 1):
    if n % i == 0:
        print(i)
```

```python
# Program to count the number of factors of a given positive number
n = int(input("Enter a positive number: "))
count = 0
for i in range(1, n + 1):
    if n % i == 0:
        count += 1
print("Number of factors:", count)
```

```python
# Program to check whether a number is Prime or not
n = int(input("Enter a number: "))
if n <= 1:
    print("Not Prime")
else:
    is_prime = True
    for i in range(2, n):
        if n % i == 0:
            is_prime = False
            break
    if is_prime:
        print("Prime")
    else:
        print("Not Prime")
```

```python
#to find HCF of two Numbers
a=int(input("enter 1st number"))
b=int(input("enter 2nd number"))
ans=0
if a<b:
    ran=a
else:
    ran=b
for i in range(1,ran):
    if a%i==0 and b%i==0:
        ans=i
print(ans)
```

```python
#to find LCM of two Numbers
a=int(input("enter 1st number"))
b=int(input("enter 2nd number"))
ans=0
if a<b:
    ran=a
else:
    ran=b
for i in range(ran,a*b+1):
    if i%a==0 and i%b==0:
        ans=i
        break
print(ans)
```
