```python
# Replace 'pi' with '3.14' recursively
def replace_pi(s):
    if len(s) < 2:
        return s
    if s[:2] == "pi":
        return "3.14" + replace_pi(s[2:])
    return s[0] + replace_pi(s[1:])
print(replace_pi("abcpipi"))
```

```python
# Binary search using recursion
def binary_search(arr, target, low, high):
    if low > high:
        return -1  
    mid = (low + high) // 2   
    if arr[mid] == target:
        return mid
    elif arr[mid] > target:
        return binary_search(arr, target, low, mid - 1)
    else:
        return binary_search(arr, target, mid + 1, high)
arr = [1, 3, 5, 7, 9, 11]
print(binary_search(arr, 7, 0, len(arr) - 1))
```
