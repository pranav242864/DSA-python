```python
# Count occurrences of a character in a string recursively

def count_char(s, ch, i=0):
    if i == len(s):
        return 0
    count = 1 if s[i] == ch else 0
    return count + count_char(s, ch, i + 1)

print(count_char("recursion", "r"))
```

```python
# Return first index of character in string recursively

def first_index(s, ch, i=0):
    if i == len(s):
        return -1
    if s[i] == ch:
        return i
    return first_index(s, ch, i + 1)

print(first_index("recursion", "s"))
```

```python
# Remove duplicate characters recursively

def remove_duplicates(s, result=""):
    if len(s) == 0:
        return result
    if s[0] not in result:
        result += s[0]
    return remove_duplicates(s[1:], result)

print(remove_duplicates("programming"))
```

```python
# Replace all occurrences of a character recursively

def replace_char(s, old, new):
    if len(s) == 0:
        return ""
    if s[0] == old:
        return new + replace_char(s[1:], old, new)
    return s[0] + replace_char(s[1:], old, new)

print(replace_char("banana", "a", "x"))
```

```python
# Remove all 'x' characters recursively

def remove_x(s):
    if len(s) == 0:
        return ""
    if s[0] == 'x':
        return remove_x(s[1:])
    return s[0] + remove_x(s[1:])

print(remove_x("axbxcxd"))
```
