```python
#Check if List is Sorted (Slicing + Recursion)
def is_sorted(arr):
    if len(arr) <= 1:
        return True
    if arr[0] > arr[1]:
        return False
    return is_sorted(arr[1:])
arr = [1, 2, 3, 4, 5]
print(is_sorted(arr))
```

```python
# Check if list is sorted using indexing + recursion
def is_sorted(arr, i=0):
    if i >= len(arr) - 1:
        return True
    if arr[i] > arr[i + 1]:
        return False
    return is_sorted(arr, i + 1)
arr = [1, 2, 3, 4, 5]
print(is_sorted(arr))
```

```python
# Return the first index of integer x present in the array (use slicing)
def first_index(arr, x):
    if len(arr) == 0:
        return -1
    if arr[0] == x:
        return 0
    small_ans = first_index(arr[1:], x)
    if small_ans == -1:
        return -1
    return small_ans + 1
arr = [1, 2, 3, 4, 3, 5]
x = 3
print(first_index(arr, x))
```

```python
# Return the first index of integer x present in the array (without using slicing)
def first_index(arr, x, i=0):
    if i == len(arr):
        return -1
    if arr[i] == x:
        return i
    return first_index(arr, x, i + 1)
arr = [1, 2, 3, 4, 3, 5]
x = 3
print(first_index(arr, x))
```

```python
# Return the last index of integer x present in the array (use recursion)
def last_index(arr, x, i=0):
    if i == len(arr):
        return -1
    small_ans = last_index(arr, x, i + 1)
    if small_ans != -1:
        return small_ans
    if arr[i] == x:
        return i
    return -1
arr = [1, 2, 3, 4, 3, 5]
x = 3
print(last_index(arr, x))
```

```python
# Return all indices of integer x present in the array (use recursion)
def all_indices(arr, x, i=0):
    if i == len(arr):
        return []
    result = all_indices(arr, x, i + 1)
    if arr[i] == x:
        return [i] + result
    return result
arr = [1, 2, 3, 4, 3, 5, 3]
x = 3
print(all_indices(arr, x))
```

```python
# Union of two arrays with duplicates (return distinct elements)
def union_arrays(a, b):
    result = []    
    for num in a:
        if num not in result:
            result.append(num)   
    for num in b:
        if num not in result:
            result.append(num)    
    return result
a = [1, 2, 2, 3, 4]
b = [3, 4, 4, 5, 6]
print(union_arrays(a, b))
```

```python
# Rotate array by one position in clockwise direction
def rotate_by_one(arr):
    if len(arr) == 0:
        return arr
    last = arr[-1]
    for i in range(len(arr) - 1, 0, -1):
        arr[i] = arr[i - 1]
    arr[0] = last
    return arr
arr = [1, 2, 3, 4, 5]
print(rotate_by_one(arr))
```

```python
# Kadane's Algorithm â€“ maximum subarray sum
def kadane(arr):
    max_sum = arr[0]
    current_sum = arr[0]    
    for i in range(1, len(arr)):
        current_sum = max(arr[i], current_sum + arr[i])
        max_sum = max(max_sum, current_sum)    
    return max_sum
arr = [-2, 1, -3, 4, -1, 2, 1, -5, 4]
print(kadane(arr))
```
