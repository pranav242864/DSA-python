```python
# print 1 to n recursively
def print_numbers(n):
    if n == 0:
        return
    print_numbers(n - 1)
    print(n)
n = int(input("enter a number"))
print_numbers(n)
```

```python
#print n to 1 recursively
def print_reverse(n):
    if n == 0:
        return
    print(n)
    print_reverse(n - 1)
n = int(input())
print_reverse(n)
```

```python
#find sum of first 10 natural no. recursively
def sum_n(n):
    if n == 1:
        return 1
    return n + sum_n(n - 1)
print(sum_n(10))
```

```python
#sun of array recursively
def sum_array(arr, n):
    if n == 0:
        return 0
    return arr[n - 1] + sum_array(arr, n - 1)
arr = [1, 2, 3, 4, 5]
print(sum_array(arr, len(arr)))
```

```python
#and arry of length n is given and an integer x you need to find if x is presssent in arry or not print true or flase
def find_element(arr, n, x):
    if n == 0:
        return False
    if arr[n - 1] == x:
        return True
    return find_element(arr, n - 1, x)
arr = [1, 3, 5, 7, 9]
x = 5
print(find_element(arr, len(arr), x))
```

```python
#find factorial 
def factorial(n):
    if n == 0 or n == 1:
        return 1
    return n * factorial(n - 1)

n = int(input())
print(factorial(n))
```

```python
# sorted list using sclising and recursion
def is_sorted(arr):
    if len(arr) <= 1:
        return True
    if arr[0] > arr[1]:
        return False
    return is_sorted(arr[1:])

arr = [1, 2, 3, 4, 5]
print(is_sorted(arr))
```
