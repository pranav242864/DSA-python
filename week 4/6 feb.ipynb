```python
# Return the last index of element x in the array (check from rear, recursion)
def last_index(arr, x, i=None):
    if i is None:
        i = len(arr) - 1
    if i < 0:
        return -1
    if arr[i] == x:
        return i
    return last_index(arr, x, i - 1)
arr = [1, 2, 3, 4, 3, 5]
x = 3
print(last_index(arr, x))
```

```python
# Find if character is present in the string recursively
def find_char(s, ch, i=0):
    if i == len(s):
        return False
    if s[i] == ch:
        return True
    return find_char(s, ch, i + 1)
s = "recursion"
ch = "s"
print(find_char(s, ch))
```
