```python
#to check if elegibal to vote
age=int(input("enter your age"))
if age >=18:
    print("eligibleto vote")
else:
    print("not eligible to vote")
```

```python
#check even or odd
num=int(input("enter a number"))
if num%2==0:
    print("even number")
else:
    print("odd number ")
```

```python
#check multiple of 5 or not
num=int(input("enter a number"))
if num%5==0:
    print(num,"multiple of 5 ")
else:
    print(num," is not a multiple of 5 ")
```

```python
#if no. is divisible by 35 or not
num=int(input("enter a number"))
if num%35==0:
    print(num,"divisible by 35 ")
else:
    print(num," is not a divisible by 35 ")
```

```python
# to check if a tringle is isoceles,equilateral or scalen
A=float(input("enter side A of a triangle "))
B=float(input("enter side B of a triangle "))
C=float(input("enter side C of a triangle "))
#to handle edge case 
if(A!=0 and B!=0 and C!=0 ) and (A>0 and B>0 and C>0):
        if A==B and A!=C :
            print("isoceles triangle")
        if A==B and A==C :
            print("equilateral triangle")
        if A!=B and A!=C :
            print("scalen triangle")
else:
    print("not a triangle")

```
